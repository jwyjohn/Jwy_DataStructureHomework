```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from math import log
matplotlib.use('TkAgg')
```

```python
sort_names = ['default_sort',
              'bubble_sort',
              'selection_sort',
              'insertion_sort',
              'binary_insertion_sort',
              'shell_sort',
              'quick_sort',
              'heap_sort',
              'merge_sort',
              'lsd_radix_sort',
              'msd_radix_sort']
res = {name:[] for name in sort_names}
```

```python
res
```

```python
lines = []
with  open("results_py.txt",'r') as f:
    lines = f.readlines()
lines[0:5]
```

```python
res = {name:[] for name in sort_names}
for l in lines:
    name, num, clock = l.strip().split(',')
    res[name].append((eval(num),eval(clock)))
```

```python
plt.style.available
```

```python
plt.style.use('default')
```

```python
dpi = 300
fig = plt.figure(figsize=(20, 9), dpi=dpi)
plt.grid(True, lw = 0.5, ls = '--', c = '.65',which='both')
plt.xlabel('N')
plt.ylabel('Time')
for i in res.items():
    tx = [j[0] for j in i[1]]
    ty = [j[1] for j in i[1]]
    plt.loglog(tx,ty,label=i[0])
plt.legend()
fig.show()
```
